```python
class Node:
    def __init__(self,data=None , next = None):
        self.data = data 
        self.next = next
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
        
    def printLL(self):
        if not self.head:
            print("No linked list is there!")
            return 
        curr = self.head
        lle = " "
        while curr:
            lle += str(curr.data)+'-->'
            curr = curr.next
        print(lle)
    
    def insert_at_beginning(self,data):
        node = Node(data,self.head)
        self.head = node
        
    def insert_at_end(self,data):
        if not self.head:
            self.head = Node(data,None)
        curr = self.head
        while curr.next:
            curr = curr.next
        curr.next = Node(data,None)
    
    def insert_values(self,data_list):
        self.head = None
        for data in data_list:
            self.insert_at_end(data)
    
    def get_length(self):
        curr= self.head
        count = 0
        while curr:
            count += 1
            curr = curr.next
        return count
    
    def remove_at(self,index):
        if index<0 or index>=self.get_length():
            print("Invalid Index !!!")
        if index==0:
            self.head = self.head.next
            return 
        count = 0 
        curr = self.head
        while curr:
            if count == index-1:
                curr.next = curr.next.next
                break
            curr = curr.next
            count+=1
    
    def insert_at(self, index , data):
        if index<0 or index>=self.get_length():
            print("Invalid Index !!!")
        if index ==0:
            self.insert_at_beginning(data)
            return
        count = 0
        curr = self.head
        while curr:
            if count == index-1:
                node = Node(data , curr.next)
                curr.next = node
            curr = curr.next
            count += 1
    
    def insert_after_value(self , data_after , data_to_insert):
        if not self.head:
            return
        if self.head.data == data_after:
            self.head.next = Node(data_to_insert,self.head.next)
        curr = self.head
        while curr:
            if curr.data == data_after:
                node = Node(data_to_insert , curr.next)
                curr.next = node
                break
            curr = curr.next
            
    def remove_by_value(self,data):
        if not self.head:
            return 
        if self.head.data  == data:
            self.head = self.head.next
        curr = self.head
        while curr.next:
            if curr.next.data == data:
                curr.next = curr.next.next
                break
            curr = curr.next
```

```python
if __name__ == '__main__':
    ll = LinkedList()
    ll.insert_values(["A","B","C","D"])
    ll.printLL()
    ll.insert_after_value("B","E")
    ll.printLL()
    ll.remove_by_value("D")
    ll.printLL()
    ll.remove_by_value("E")
    ll.printLL()
    ll.remove_by_value("A")
    ll.remove_by_value("C")
    ll.remove_by_value("E")
    ll.remove_by_value("C")
    ll.printLL()
    
```
